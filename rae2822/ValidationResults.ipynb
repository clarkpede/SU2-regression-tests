RAE 2822 Transonic Airfoil
==========================

+ Transonic, with a shock wave
+ 2D

See https://www.grc.nasa.gov/WWW/wind/valid/raetaf/raetaf.html

```python
import numpy as np
import matplotlib.pyplot as plt
import sys
sys.path.append('../')
from regression_plots import surfaceplot, histplot
plt.style.use(('seaborn-whitegrid', 'seaborn-notebook'))
```

```python
experiment = np.genfromtxt("RAE2822_case9_experimental.plt", skip_header=2, names=('x', 'Cp'), delimiter=",")
```

## SST

```python
commits = ["4ccec15", "b554dd4"]
formats = ["-", "--"]
```

```python
fig, ax = surfaceplot("SST", commits, formats)
ax[0].plot(experiment["x"], experiment["Cp"], "o", label="exp")
ax[0].legend(loc="best")
fig.tight_layout()
fig.savefig("RAE2822_sst_surface.png")
```

```python
fig, ax = histplot("SST", commits, formats)
fig.savefig("RAE2822_sst_history.png")
```
